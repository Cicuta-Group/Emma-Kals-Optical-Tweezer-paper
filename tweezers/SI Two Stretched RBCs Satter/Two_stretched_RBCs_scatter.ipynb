```python

filename = 'Multiple attachments for tweezer paper.xlsx'

df_twoforces_all = pd.read_excel(
    filename, 
    header=0,
    #usecols='A:G',
   # userows='1:12',
    index_col=0,
   # nrows=ROW_COUNT,
)
df_twoforces_all.index = df_twoforces_all.index.str.strip() # remove whitespace

#df_egresstime_raw ["strain"] = df_egresstime_raw.index
#df_egresstime_raw['Strain'] = df_egresstime_raw['strain'].str.split(" ").str[0]

print(df_twoforces_all.columns)

df_twoforces_all # display dataframe
```

```python
#This sets what condition is plotted 
conditions_to_plot = 'NF54'

df_twoforces_plot = df_twoforces_all[df_twoforces_all['Strain'] == conditions_to_plot]

print(df_twoforces_plot.columns)

df_twoforces_plot # display dataframe
```

```python
plt.figure()
plt.gcf().set_size_inches(12,10)
plt.gcf().set_dpi(300)

sns.set_theme(#context='notebook', 
            #style='ticks', 
            style='darkgrid', 
            font='arial', 
            font_scale=3.5) 
            #   color_codes=True, 
            #   rc=None)
sns.set_style(rc = {'axes.facecolor': '#F5F5F9'})
#sns.set_style(rc = {'axes.facecolor': '#FBF9F6'})


### calculate the coefficent of variance 

x_column = 'Force x (pN)' 
y_column = 'Force y (pN)' 

# Extract data from DataFrame columns
x_data = df_twoforces_plot[x_column].values
y_data = df_twoforces_plot[y_column].values

# Calculate correlation coefficient
correlation_matrix = np.corrcoef(x_data, y_data)
correlation_coefficient = correlation_matrix[0, 1]

 # Calculate R-squared
r_squared = correlation_matrix[0, 1] ** 2


legend_label = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f} \nn={len(df_twoforces_plot)}"
#legend_label = f"R-squared = {r_squared:.3f}"

print(f"Correlation Coefficient: {correlation_coefficient}")

#label = f'Fit: y = {slope:.2f} * log10(x) + {intercept:.2f}\nR-squared = {r_squared:.2f}\nEfficiency = {efficiency:.0f}%'




sns.regplot(
    data=df_twoforces_plot,
    x="Force x (pN)", 
    y="Force y (pN)",
    scatter=False,
    line_kws={'color': '.7'},
    fit_reg=True,
    ci=None,
    )


sns.scatterplot(
    data=df_twoforces_plot, 
    x="Force x (pN)", 
    y="Force y (pN)", 
    hue=df_twoforces_plot.Strain,
    palette=colors_light,
    edgecolor=[colors_dark.get(s, 'r') for s in df_twoforces_plot.Strain],
    #facecolor=[colors_light.get(s, 'r') for s in df_twoforces_plot.Strain],
    s=60,
    marker='o',
    linewidth=3,
    label=legend_label,
)



plt.xlim((0,90))
plt.ylim((0,90))
plt.legend(markerscale=0)
plt.title(conditions_to_plot)

#sns.move_legend(plt.gca(), "upper left", bbox_to_anchor=(1, 1))

plt.rcParams['svg.fonttype'] = 'none'
plt.savefig(f'graphs/SupFig1b_two_forces_{conditions_to_plot}.svg', bbox_inches='tight')
```
