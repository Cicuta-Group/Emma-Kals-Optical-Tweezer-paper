```python

# filename = r'C:/Users/Emma Jones/OneDrive - University of Cambridge/Documents/My work/cambridge/Year 2/Rayner lab/Data/Optical tweezers/Testing what effects detachment force/Scatter_data_optical_tweezers_summary.xlsx'
filename = r'Scatter_data_optical_tweezers_summary.xlsx'

df_OT_raw = pd.read_excel(
    filename, 
    header=0,
   # usecols='A:CH',
   # userows='1:12',
    index_col=1,
   # nrows=ROW_COUNT,
)
df_OT_raw.index = df_OT_raw.index.str.strip() # remove whitespace

df_OT_raw ["strain"] = df_OT_raw.index
df_OT_raw['Strain'] = df_OT_raw['strain'].str.split(" ").str[0]


print(df_OT_raw.columns)

df_OT_raw # display dataframe
```

```python

plt.figure()
plt.gcf().set_size_inches(12,9)
plt.gcf().set_dpi(300)

# #Presnetation theme 
# sns.set_theme(context='notebook', style='darkgrid', font='Candara', font_scale=2, color_codes=True, rc=None, )
# #sns.set_style(rc = {'axes.facecolor': '#FBF9F6'})
# sns.set_style(rc = {'axes.facecolor': '#F5F5F9'})

#Paper theme 
sns.set_theme(#context='notebook', 
            #style='ticks', 
            style='darkgrid', 
            font='arial', 
            font_scale=3.5,
            ) 
            #   color_codes=True, 
            #   rc=None)
sns.set_style(rc = {'axes.facecolor': '#F5F5F9'})


#Figure 6
# ###Variables to be plotted 
# #Forces vs static invasion rate 
# x_column = "Detachment force (pN)"  
# y_column = "Invasion efficiency relative to NF54 (%)"

# x_column_err = "Attachment force SEM"
# y_column_err = "SEM of invasion efficiency relative to NF54"



# ###Variables to be plotted 
# #Frequency vs static invasion rate 
# x_column = "Attachment frequency (%) " 
# y_column = "Invasion efficiency  relative to NF54 (%)"

# x_column_err = "SEM of atatched per egress"
# y_column_err = "SEM of invasion efficiency relative to NF54" 

#Figure 9 - knockouts 

# ###Variables to be plotted 
# #Forces vs static invasion rate 
# x_column = "Detachment force (pN)"  
# y_column = "Mean Growth rate (PEMR)"

# x_column_err = "Attachment force SEM"
# y_column_err = "SEM Invasion rate in static "


#Frequency vs static invasion rate 
x_column = "Attachment frequency (%) " 
y_column = "Mean Growth rate (PEMR)"

x_column_err = "SEM of atatched per egress"
y_column_err = "SEM Invasion rate in static " 

# Graphs 
for _, row in df_OT_raw.iterrows():
    strain = row['Strain']
    color = colors_dark.get(strain, 'black')
    print(color)

    plt.errorbar(
        data=row,
        x=x_column,
        y=y_column,
        xerr=x_column_err,
        yerr=y_column_err,
        fmt=".",
        color=color,
        ecolor=color,
        label=None,
        # elinewidth=1,
        capsize=5,
        # zorder=1,
        markersize=0,
    )

sns.regplot(
    x=x_column,
    y=y_column,
    ci=None,
    data=df_OT_raw, 
    scatter=False,
    line_kws={'color': '.7'},
)


# #Fig 1 
# NotLate = ['3D7', 'NF54', 'Anti-GYPA', 'Neuraminidase	', 'Anti-GYPC']

# df_NotLate = df_OT_raw.query('Strain == @NotLate')

# sns.regplot(
#     x=x_column,
#     y=y_column,
#     ci=None,
#     data=df_NotLate, 
#     scatter=False,
#     line_kws={'color': 'r'},
# )

sns.scatterplot(
    data=df_OT_raw, 
    x=x_column, 
    y=y_column,
    hue='Strain',
    palette=colors_dark,
    # edgecolor='.5', 
    s=100,
    style="Assay ",
)

### calculate the coefficent of variance (now above)

# x_column = "Mean detachment force (pN)" 
# y_column = "Invasion efficeny  relative to NF54 static (%)"

# Extract data from DataFrame columns
x_data = df_OT_raw[x_column].values
y_data = df_OT_raw[y_column].values

# #For part 1 red 
# # Extract data from DataFrame columns
# x_data = df_NotLate[x_column].values
# y_data = df_NotLate[y_column].values

# Calculate correlation coefficient
correlation_matrix = np.corrcoef(x_data, y_data)
correlation_coefficient = correlation_matrix[0, 1]

 # Calculate R-squared
r_squared = correlation_matrix[0, 1] ** 2


#legend_label = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f}"
#legend_label = f"R-squared = {r_squared:.3f}"
legend_label = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f} \nn={len(df_OT_raw)}"

print(f"Correlation Coefficient: {correlation_coefficient}")


label_info = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f} \nn={len(df_OT_raw)}"
plt.annotate(label_info, xy=(0.05, -0.5), xycoords='axes fraction', fontsize=40)

plt.xlim((0,40))

#Fig S6
plt.ylim((0,300))

#Fig S9
plt.ylim((0,10))

plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

# this allows the text to be editable in affinitity (it rescales the text)
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig(f'graphs/Scatter_plots_OT_S8b_growth.svg', bbox_inches='tight')
```
