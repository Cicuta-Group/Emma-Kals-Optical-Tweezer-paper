```python

filename = 'Comparision of time post egress and detachment force.xlsx'

df_egresstime_all_raw = pd.read_excel(
    filename, 
    header=0,
   # usecols='A:CH',
   # userows='1:12',
    index_col=0,
   # nrows=ROW_COUNT,
)
df_egresstime_all_raw.index = df_egresstime_all_raw.index.str.strip() # remove whitespace

#df_egresstime_raw ["strain"] = df_egresstime_raw.index
#df_egresstime_raw['Strain'] = df_egresstime_raw['strain'].str.split(" ").str[0]

print(df_egresstime_all_raw.columns)

df_egresstime_all_raw # display dataframe
```

```python
conditions_to_plot = '3D7'

df_egresstime_raw = df_egresstime_all_raw[df_egresstime_all_raw['Strain'] == conditions_to_plot]

print(df_egresstime_raw.columns)

df_egresstime_raw # display dataframe
```

```python
plt.figure()
plt.gcf().set_size_inches(12,9)
plt.gcf().set_dpi(300)

sns.set_theme(#context='notebook', 
            #style='ticks', 
            style='darkgrid', 
            font='arial', 
            font_scale=3.5) 
            #   color_codes=True, 
            #   rc=None)
sns.set_style(rc = {'axes.facecolor': '#F5F5F9'})
#sns.set_style(rc = {'axes.facecolor': '#FBF9F6'})


### calculate the coefficent of variance 

x_column = 'Time after egress (sec)  ' 
y_column = 'Force (pN)' 

# Extract data from DataFrame columns
x_data = df_egresstime_raw[x_column].values
y_data = df_egresstime_raw[y_column].values

# Calculate correlation coefficient
correlation_matrix = np.corrcoef(x_data, y_data)
correlation_coefficient = correlation_matrix[0, 1]

 # Calculate R-squared
r_squared = correlation_matrix[0, 1] ** 2


#legend_label = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f}"
#legend_label = f"R-squared = {r_squared:.3f}"
legend_label = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f} \nn={len(df_egresstime_raw)}"

print(f"Correlation Coefficient: {correlation_coefficient}")

#label = f'Fit: y = {slope:.2f} * log10(x) + {intercept:.2f}\nR-squared = {r_squared:.2f}\nEfficiency = {efficiency:.0f}%'




sns.regplot(
    data=df_egresstime_raw,
    x="Time after egress (sec)  ", 
    y="Force (pN)",
    scatter=False,
    line_kws={'color': '.7'},
    fit_reg=True,
    ci=None,
    )


sns.scatterplot(
    data=df_egresstime_raw, 
    x="Time after egress (sec)  ", 
    y="Force (pN)", 
    # #NF54
    # edgecolor='#043546',
    # facecolor='#829AA3',
    # #3D7
    # edgecolor='#818897',
    # facecolor='#B9BDC5',
    #other
    # edgecolor='darkblue',
    # facecolor='darkblue',
    hue=df_egresstime_raw.Strain,
    palette=colors_light,
    edgecolor=[colors_dark.get(s, 'r') for s in df_egresstime_raw.Strain],
    s=100,
    marker='o',
    linewidth=2,
    #label=legend_label,
    legend=False,
)



plt.xlim((0,220))
plt.ylim((0,90))
#plt.legend()
plt.title(conditions_to_plot)
#sns.move_legend(plt.gca(), "upper left", bbox_to_anchor=(1, 1))

label_info = f"Correlation coefficient: {correlation_coefficient:.3f}\nR-squared = {r_squared:.3f} \nn={len(df_egresstime_raw)}"
plt.annotate(label_info, xy=(0.05, -0.5), xycoords='axes fraction', fontsize=40)

plt.rcParams['svg.fonttype'] = 'none'
plt.savefig(f'Graphs/SupFig1d_egress time vs force_{conditions_to_plot}.svg', bbox_inches='tight')
```
