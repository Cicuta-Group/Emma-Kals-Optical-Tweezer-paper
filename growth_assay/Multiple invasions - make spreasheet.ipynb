```python
## convert mutliple sheets to one and transpose rows

import os
import pandas as pd
from natsort import natsorted

# GA2
path = r'C:\Users\Emma Jones\OneDrive - University of Cambridge\Documents\My work\cambridge\Year 2\Rayner lab\Data\Growth assay\Growth assay 2\Multiple invasion spreadsheets'
output_file = 'GA2 Multiple invasion summary.xlsx'
mapping = GA2_remapping

# GA2 extra
# path = r'C:\Users\Emma Jones\OneDrive - University of Cambridge\Documents\My work\cambridge\Year 2\Rayner lab\Data\Growth assay\Extra plates GA2 and GA3'
# output_file = 'GA2 extra Multiple invasion summary.xlsx'
# mapping = GA2_extra_remapping

# GA3
# path = r'C:\Users\Emma Jones\OneDrive - University of Cambridge\Documents\My work\cambridge\Year 2\Rayner lab\Data\Growth assay\Growth assay 3\Multiple invasion spreadsheets'
# output_file = 'GA3 Multiple invasion summary.xlsx'
# mapping = GA3_remapping

files = [f for f in os.listdir(path) if f.endswith('.xls')]
files = sorted(files)

filepaths = [os.path.join(path, f) for f in files]

def df_from_file(filepath):
    
    day = os.path.basename(filepath).split(' ')[3]
    print('Anaysing data for day', day)
    
    df = pd.read_excel(filepath)
    df['Well'] = df['Name'].str.split('.').str[0]
    df['Property'] = df['Name'].str.split('.').str[-1].str.split('/').str[-1]

    # Melt the DataFrame
    melted_df = pd.melt(df, id_vars=['Well', 'Property'], value_vars=['#Cells'], var_name='Variable', value_name='Value')
    df2 = melted_df.pivot_table(index='Well', columns='Property', values='Value', aggfunc='first').reset_index()

    mapping_day = mapping.get(int(day), {})
    print(mapping_day)

    mapping_day = {number_to_well(k): number_to_well(v) for k,v in mapping_day.items()} 
    print(mapping_day)

    def update_name(well):
        return mapping_day.get(well, well)

    df3 = df2.copy()
    df3['Well0'] = df3['Well']
    df3['Well'] = df3['Well'].apply(update_name)
    df3['Day'] = day
    df3 = df3.dropna(subset='Well')

    sorted_wells = natsorted(df3['Well'])
    df3 = df3.set_index('Well').loc[sorted_wells].reset_index()

    return df3

dfs = [df_from_file(f) for f in filepaths]

# df = pd.concat(dfs)
# df.to_excel(output_file)
# df
```
